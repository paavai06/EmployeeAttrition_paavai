```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head(10)
```

```python
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
target = 'Attrition'
print(f"Target column: {target}")
```

```python
attrition_counts = df[target].value_counts()
print(attrition_counts)
#print(f"Attrition rate: {attrition_rate:.2f}%")
```

```python
attrition_rate = attrition_counts['Yes']/len(df)*100
print(f"Attrition rate: {attrition_rate:.2f}%")
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
print(f"Numeric columns ({len(numeric_cols)}): {list(numeric_cols)}")
print(f"Categorical columns ({len(categorical_cols)}): {list(categorical_cols)}")
```

```python
print(df.isnull().sum())
```

```python
cols_to_drop = ['EmployeeNumber','Over18','StandardHours']
df_clean = df.drop(columns = cols_to_drop, errors='ignore')
print(f"Columns after dropping: {df_clean.columns.tolist()}")
```

```python
df_clean['Attrition'] = df_clean['Attrition'].map({'Yes':1,'No':0})
```

```python
categorical_cols = df_clean.select_dtypes(include=['object']).columns
print(f"Categorical column to encode: {list(categorical_cols)}")
```

```python
df_encoded = pd.get_dummies(df_clean,columns = categorical_cols,drop_first=True)
print(f"Shape after encoding: {df_encoded.shape}")
```

```python
X = df_encoded.drop('Attrition',axis=1)
y = df_encoded['Attrition']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled,columns = X.columns)
print("Preprocessing complete.")
```

```python
dept_attrition = df.groupby('Department')['Attrition'].apply(lambda x: (x=='Yes').mean()*100)
print(dept_attrition.sort_values(ascending=False))
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=dept_attrition.index, y=dept_attrition.values)
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score,confusion_matrix,roc_curve
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state = 42,stratify=y)
```

```python
models ={ 'Logistic Regression' : LogisticRegression(class_weight ='balanced',random_state = 42,max_iter=1000),
         'Random Forest' : RandomForestClassifier(class_weight ='balanced', random_state =42),
         'Gradient Boosting': GradientBoostingClassifier(random_state=42)}
         
```

```python
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced',classes = np.unique(y_train),y=y_train)
sample_weights = np.array([class_weights[1] if i==1 else class_weights[0] for i in y_train])
```

```python
results = {}
```

```python
for name, model in models.items():
    if name == 'Gradient Boosting':
        model.fit(X_train,y_train,sample_weight=sample_weights)
    else:
        model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]
    report = classification_report(y_test,y_pred,output_dict=True)
    auc = roc_auc_score(y_test,y_proba)
    cm = confusion_matrix(y_test,y_pred)
    results[name] = {
        'precision': report['1']['precision'],
        'recall': report['1']['recall'],
        'f1': report['1']['f1-score'],
        'roc_auc': auc,
        'confusion_matrix': cm,
        'model': model,
        'y_pred': y_pred,
        'y_proba': y_proba
    }
comparison_df = pd.DataFrame({
    model: {metric: results[model][metric] for metric in ['precision', 'recall', 'f1', 'roc_auc']}
    for model in results
}).T
print("Model Comparison Table:")
print(comparison_df)

# Identify best model (based on F1-score or ROC-AUC)
best_model_name = comparison_df['f1'].idxmax()
print(f"\nBest model based on F1-score: {best_model_name}")

# Confusion matrices
for name, res in results.items():
    print(f"\n{name} Confusion Matrix:")
    print(res['confusion_matrix'])

```

```python
best_model = results[best_model_name]['model']
if hasattr(best_model,'feature_importances_'):
    importances = best_model.feature_importances_
    features_names = X.columns
    feat_imp = pd.Series(importances,index=feature_names).sort_values(ascending = False)
    print("\nTop 10 most importances features:")
    print(feat_imp.head(10))
elif hasattr(best_model,'coef_'):
    coefs = best_model.coef_[0]
    feat_imp = pd.Series(np.abs(coefs),index=X.columns).sort_values(ascending = False)
    print("\nTop 10 most important features (absolute coefficient):")
    print(feat_imp.head(10))
else:
    print("No feature importance available for this model.")
```
